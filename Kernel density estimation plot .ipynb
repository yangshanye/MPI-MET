```python
def MBE(y_true, y_pred):
    '''
    Parameters:
        y_true (array): Array of observed values
        y_pred (array): Array of prediction values

    Returns:
        mbe (float): Biais score
    '''
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_true = y_true.reshape(len(y_true),1)
    y_pred = y_pred.reshape(len(y_pred),1)   
    diff = (y_true-y_pred)
    mbe = diff.mean()
    print('MBE = ', mbe)
```

```python
import sympy as sp
import math

# 定义变量
D = sp.symbols('D')

# 定义函数P
P = 38.045 * D ** -1.196

# 计算x（P关于D的积分从？到？微米）
x = sp.integrate(P, (D, 5.9, 5000))
# 微塑料个数
N = 782
# 计算dP/dD
dP_dD_abs = sp.Abs(sp.diff(P, D))

# 定义积分表达式
integrand = dP_dD_abs / x * N * math.pi * D**3 / 6 * 1e-6

# 计算积分
result = sp.integrate(integrand, (D, 0.5, 70))

print(f"The value of x is: {x}")
print(f"mass of microplastic from 5 to 5000 is: {result}")
```
