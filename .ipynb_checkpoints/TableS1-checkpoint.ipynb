```python
import pandas as pd

for j in range(6):
    # dep kg/grid
    dep = MP[11][drys[j]][0,:,:]+MP[11][wets[j]][0,:,:]
    
    # dep kg/s
    dep = dep/180*3600*24*365
    
    # lon-180
    dep.coords['lon'] = (dep.coords['lon'] + 180) % 360 - 180
    dep = dep.sortby(dep.lon)
    
    world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
    
    locals()['df_dep'+str(j+1)]= pd.DataFrame({'Name': [], 
                  'Sum': []})

    for i in range(177):
    
        conus=world.iloc[[i]]
        conus.geometry
        pr=dep
        pr.rio.set_spatial_dims("lon", "lat", inplace=True)
        pr.rio.write_crs("epsg:4326", inplace=True)
        clipped1 = pr.rio.clip(conus.geometry.apply(mapping), conus.crs, drop=False, invert=False)     # clips everything - all nan
        sum_MP=clipped1.sum()
        locals()['df_dep'+str(j+1)]=locals()['df_dep'+str(j+1)].append({'Name': str(world.iloc[i,2]),
                  'Sum': float(sum_MP)},ignore_index=True)
```

```python
# sort+reverse+reorder_index
for i in range(6):
    locals()['df_dep'+str(i+1)]=locals()['df_dep'+str(i+1)].sort_values(by=['Sum']).iloc[::-1].reset_index(drop=True)
```
