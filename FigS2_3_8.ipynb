```python
import matplotlib.pyplot as plt

# Data for Location 1
location1 = ['Woman', 'Man']
values_location1 = [4, 5]

# Data for Location 2
location2 = ['Woman', 'Man']
values_location2 = [8, 18]

# Set the positions of the bars on the x-axis
x = range(len(location1))

# Plot the bars
plt.bar(x, values_location1, width=0.4, label='Location 1')
plt.bar([i + 0.4 for i in x], values_location2, width=0.4, label='Location 2')

# Set the x-axis labels
plt.xticks([i + 0.2 for i in x], location1)

# Set the y-axis label
plt.ylabel('Count')

# Set the chart title
plt.title('Bar Chart - Location 1 vs Location 2')

# Add a legend
plt.legend()

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for particle size
particle_size = [0.5, 1.5, 5, 10, 35, 70]

# Data for mass concentration (μg/m3)
mass_concentration_range = [
    [3.8e-4, 2.0e-2],
    [1.6e-3, 0.2],
    [1.2e-11, 4.0e-1],
    [1.2e-11, 1.2e-1],
    [1.2e-11, 1.9e-1],
    [1.2e-11, 1.0e-1]
]

# Data for number concentration (MP/m3)
number_concentration_range = [
    [5.8e3, 3.1e5],
    [9.3e2, 1.0e5],
    [1.9e-7, 6.3e3],
    [2.3e-8, 4.1e3],
    [5.3e-10, 86.3],
    [6.5e-11, 5.7]
]

# Plotting figure 'a' for mass concentration range
fig_a, ax_a = plt.subplots()
ax_a.set_xlabel('Microplastic diameter (μm)')
ax_a.set_ylabel(r'Mass Concentration (μg/m$^{3}$)')
#ax_a.set_title('Mass Concentration Range (Plot A)')

# Convert the range data to arrays for plotting
mass_concentration_lower = np.array([range[0] for range in mass_concentration_range])
mass_concentration_upper = np.array([range[1] for range in mass_concentration_range])
x = np.arange(len(particle_size))

# Plot the range as bar plots
ax_a.bar(x, mass_concentration_upper - mass_concentration_lower, bottom=mass_concentration_lower, width=0.4, align='center')
ax_a.set_xticks(x)
ax_a.set_xticklabels(particle_size)
ax_a.set_yscale('log')

# Set y-axis limits to include the entire range of bars
ax_a.set_ylim([min(mass_concentration_lower) * 0.8, max(mass_concentration_upper) * 1.2])

# Mark the plot as 'a'
ax_a.text(0.05, 0.95, 'a', transform=ax_a.transAxes, fontsize=18, fontweight='bold', va='top', ha='right')

plt.gcf().set_dpi(300)


# Plotting figure 'b' for number concentration range
fig_b, ax_b = plt.subplots()
ax_b.set_xlabel('Microplastic diameter (μm)')
ax_b.set_ylabel('Number Concentration (MP/m$^{3}$)')
#ax_b.set_title('Number Concentration Range (Plot B)')

# Convert the range data to arrays for plotting
number_concentration_lower = np.array([range[0] for range in number_concentration_range])
number_concentration_upper = np.array([range[1] for range in number_concentration_range])

# Plot the range as bar plots
ax_b.bar(x, number_concentration_upper - number_concentration_lower, bottom=number_concentration_lower, width=0.4, align='center')
ax_b.set_xticks(x)
ax_b.set_xticklabels(particle_size)
ax_b.set_yscale('log')

# Set y-axis limits to include the entire range of bars
ax_b.set_ylim([min(number_concentration_lower) * 0.8, 1e8])

# Mark the plot as 'b'
ax_b.text(0.05, 0.95, 'b', transform=ax_b.transAxes, fontsize=18, fontweight='bold', va='top', ha='right')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.4)

plt.gcf().set_dpi(300)

# Display the plots
plt.show()
```
